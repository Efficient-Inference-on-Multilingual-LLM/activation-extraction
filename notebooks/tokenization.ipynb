```python
import pandas as pd
import numpy as np
import torch
import os
import json
from glob import glob
import matplotlib.pyplot as plt
os.chdir('..')
```

```python
from transformers import AutoTokenizer
from src.utils.const import LANGCODE2LANGNAME, LANGNAME2LANGCODE, MODEL2HIDDEN_SIZE, MODEL2NUM_LAYERS, EXP2_CONFIG, EXP3_CONFIG, EXP4_CONFIG, MODEL2HF_NAME
from datasets import load_dataset
from dotenv import load_dotenv
load_dotenv()
```

```python
from tqdm import tqdm
```

```python
languages = []
for family, langs in EXP4_CONFIG['languages'].items():
	languages.extend(langs)
languages = [LANGNAME2LANGCODE[lang] for lang in languages]
```

```python
dataset_per_lang = {}
for lang in languages:
    dataset_per_lang[lang] = load_dataset("openlanguagedata/flores_plus", lang, split='dev', cache_dir=os.getenv("HF_CACHE_DIR"))
```

```python
model_names = glob('outputs_silhouette/exp4/next_token/dev/*')
model_names = [mn.split('/')[-1] for mn in model_names]
model_names.sort()
print(model_names)
model_name = model_names[1]
```

```python
for model_name in model_names:
	tokenizer = AutoTokenizer.from_pretrained(f'{MODEL2HF_NAME[model_name]}/{model_name}', cache_dir=os.getenv("HF_CACHE_DIR"))
	print(f"Tokenizer vocab size for {model_name}: {tokenizer.vocab_size}")

	# Initialize similarity matrix with size languages x languages (numpy)
	similarity_matrix = np.zeros((len(dataset_per_lang), len(dataset_per_lang)))

	# Pre-tokenize all datasets once (major optimization!)
	tokenized_datasets = {}
	print("Pre-tokenizing all datasets...")
	for lang in tqdm(dataset_per_lang.keys(), desc="Tokenizing languages"):
		tokenized_datasets[lang] = tokenizer(
			list(dataset_per_lang[lang]['text']), 
			padding=False, 
			truncation=False, 
			return_tensors=None
		)['input_ids']

	# Calculate similarity using pre-tokenized data
	for i, lang1 in tqdm(enumerate(dataset_per_lang.keys()), total=len(dataset_per_lang.keys()), desc="Calculating similarity matrix"):
		tokenized_outputs1 = tokenized_datasets[lang1]
		
		for j, lang2 in enumerate(list(dataset_per_lang.keys())[i+1:], start=i+1):
			tokenized_outputs2 = tokenized_datasets[lang2]
			
			# Vectorized calculation using list comprehension
			similarities = [
				len(set(tok1) & set(tok2)) / len(set(tok1) | set(tok2))
				for tok1, tok2 in zip(tokenized_outputs1, tokenized_outputs2)
			]
			
			average_percentage = np.mean(similarities)
			similarity_matrix[i, j] = average_percentage
			similarity_matrix[j, i] = average_percentage

	exp_id = 'exp4'
	split = 'dev'
	output_path = f'outputs_tokenization/{exp_id}/{split}/{model_name}/'
	os.makedirs(output_path, exist_ok=True)
	np.savez_compressed(f'{output_path}/jaccard_similarity.npz', similarity_matrix)
```

```python
for model_name in model_names:
	output_path = f'outputs_tokenization/exp4/dev/{model_name}'
	similarity_matrix = np.load(f'{output_path}/jaccard_similarity.npz')['arr_0']
	print(f"Jaccard Similarity Matrix for {model_name}:")
	
	# Create heatmap
	plt.figure(figsize=(12, 10))
	plt.imshow(similarity_matrix, cmap='viridis', interpolation='nearest')
	plt.colorbar(label='Jaccard Similarity')
	plt.xticks(ticks=np.arange(len(languages)), labels=[LANGCODE2LANGNAME[lang] for lang in languages], rotation=90)
	plt.yticks(ticks=np.arange(len(languages)), labels=[LANGCODE2LANGNAME[lang] for lang in languages])

	# Reduce the ytick font size for better readability
	plt.yticks(fontsize=8)
	plt.xticks(fontsize=8)
	plt.title(f'Jaccard Similarity Tokenization Heatmap for Flores+ (200 instances) - {model_name}')
	plt.tight_layout()
	plt.savefig(f'{output_path}/jaccard_similarity_heatmap.png', dpi=300)
	plt.close()
```
